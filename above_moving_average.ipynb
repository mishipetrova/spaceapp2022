```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('alaska_temp.csv')
df = df.set_index('Year')
```

```python
df['rolling_avg_10'] = df.rolling(10).mean()
```

```python
df = df.dropna()
df = df.loc[:2016]
# df
df.head()
```

```python
fig, ax = plt.subplots(figsize=(7,5))
temp, = ax.plot(df.avg_temp_c)
roll_temp,  = ax.plot(df.rolling_avg_10)
ax.set_title('Yearly temperature variation in Alaska (1961-2016)')
ax.set_xlabel('Year')
ax.set_ylabel('Temperatures (C)')
ax.legend([temp, roll_temp], ['Mean temperature', '10-year moving average temperature'])
ax.annotate('Source: ACIS (Applied Climate Information System)',
            xy = (0.4, -0.1),
            xycoords='axes fraction',
            ha='right',
            va="center",
            fontsize=10);
plt.savefig('yearly_alaska_temp_variation.png')
```
